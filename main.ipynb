```python
LABEL_MAP = {
    "hem": 0,   # Healthy
    "all": 1    # Leukemia
}
```

```python
def get_transforms(img_size=224, augment=False):
    if augment:
        return transforms.Compose([
            transforms.Resize((img_size, img_size)),
            transforms.RandomHorizontalFlip(),
            transforms.RandomVerticalFlip(),
            transforms.RandomRotation(15),
            transforms.ColorJitter(brightness=0.1, contrast=0.1),
            transforms.ToTensor(),
            transforms.Normalize(
                mean=[0.485, 0.456, 0.406],
                std=[0.229, 0.224, 0.225]
            )
        ])
    else:
        return transforms.Compose([
            transforms.Resize((img_size, img_size)),
            transforms.ToTensor(),
            transforms.Normalize(
                mean=[0.485, 0.456, 0.406],
                std=[0.229, 0.224, 0.225]
            )
        ])
```

```python
class LeukemiaTrainDataset(Dataset):
    def __init__(self, fold_dirs, transform=None):
        self.samples = []
        self.transform = transform
        for fold_dir in fold_dirs:
            for class_name in ["hem", "all"]:
                class_dir = os.path.join(fold_dir, class_name)
                label = LABEL_MAP[class_name]
                for img_name in os.listdir(class_dir):
                    img_path = os.path.join(class_dir, img_name)
                    self.samples.append((img_path, label))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_path, label = self.samples[idx]
        image = Image.open(img_path).convert("RGB")
        if self.transform:
            image = self.transform(image)
        return image, torch.tensor(label, dtype=torch.long)
```

```python
class LeukemiaTestDataset(Dataset):
    def __init__(self, img_dir, csv_file=None, transform=None):
        self.img_dir = img_dir
        self.transform = transform
        self.csv_file = csv_file

        if csv_file is not None:
            self.df = pd.read_csv(csv_file)
            self.has_labels = True
        else:
            self.images = sorted(os.listdir(img_dir))
            self.has_labels = False

    def __len__(self):
        return len(self.df) if self.has_labels else len(self.images)

    def __getitem__(self, idx):
        if self.has_labels:
            img_name = self.df.iloc[idx]["new_names"]
            label = int(self.df.iloc[idx]["labels"])
        else:
            img_name = self.images[idx]
            label = -1  # dummy

        img_path = os.path.join(self.img_dir, img_name)
        image = Image.open(img_path).convert("RGB")

        if self.transform:
            image = self.transform(image)

        return image, torch.tensor(label, dtype=torch.long)
```

```python
class FocalLoss(nn.Module):
    def __init__(self, alpha=1, gamma=2):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.ce = nn.CrossEntropyLoss(reduction="none")

    def forward(self, logits, targets):
        ce_loss = self.ce(logits, targets)
        pt = torch.exp(-ce_loss)
        focal_loss = self.alpha * (1 - pt) ** self.gamma * ce_loss
        return focal_loss.mean()
```

```python
import torchvision.models as models
import torch.nn as nn

def get_backbone(backbone_name="resnet18", pretrained=True):
    """
    Returns a CNN backbone and its output channels (feature_dim)
    The backbone outputs a feature map [B, C, H, W] for Transformer projection
    """
    if backbone_name == "resnet18":
        model = models.resnet18(pretrained=pretrained)
        feature_dim = 512
        model.fc = nn.Identity()  # remove classifier

    elif backbone_name == "resnet50":
        model = models.resnet50(pretrained=pretrained)
        feature_dim = 2048
        model.fc = nn.Identity()

    elif backbone_name == "efficientnet_b0":
        model = models.efficientnet_b0(pretrained=pretrained)
        feature_dim = 1280
        model = model.features  # remove classifier, keep feature map

    else:
        raise ValueError("Unsupported backbone")

    return model, feature_dim
```

```python
import torch
import torch.nn as nn
from torchvision import models

class CNNTransformerClassifier(nn.Module):
    def __init__(self, backbone_name="resnet18", num_classes=2,
                 d_model=256, nhead=8, num_layers=2, pretrained=True):
        super().__init__()
        # ----------------- BACKBONE -----------------
        if backbone_name == "resnet18":
            self.backbone = models.resnet18(pretrained=pretrained)
            feature_dim = self.backbone.fc.in_features
            self.backbone.fc = nn.Identity()
        elif backbone_name == "resnet50":
            self.backbone = models.resnet50(pretrained=pretrained)
            feature_dim = self.backbone.fc.in_features
            self.backbone.fc = nn.Identity()
        elif backbone_name == "efficientnet_b0":
            self.backbone = models.efficientnet_b0(pretrained=pretrained)
            feature_dim = self.backbone.classifier[1].in_features
            self.backbone.classifier = nn.Identity()
        else:
            raise ValueError("Unsupported backbone")

        # ----------------- TRANSFORMER ENCODER -----------------
        self.transformer = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(d_model=d_model, nhead=nhead, batch_first=True),
            num_layers=num_layers
        )

        # Project CNN features to transformer dimension
        self.feature_proj = nn.Linear(feature_dim, d_model)

        # ----------------- CLASSIFIER -----------------
        self.classifier = nn.Sequential(
            nn.Linear(d_model, 256),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        features = self.backbone(x)  # [B, C]
        features = self.feature_proj(features)  # [B, d_model]
        features = features.unsqueeze(1)  # [B, 1, d_model] as seq_len=1
        features = self.transformer(features)  # [B, 1, d_model]
        features = features.squeeze(1)  # [B, d_model]
        out = self.classifier(features)
        return out
```

```python
# ------------------------- STEP 7: FOLD-AWARE LOADERS -------------------------
def get_fold_loaders(train_folds, val_folds, img_size, batch_size):
    train_dataset = LeukemiaTrainDataset(fold_dirs=train_folds, transform=get_transforms(img_size, augment=True))
    val_dataset = LeukemiaTrainDataset(fold_dirs=val_folds, transform=get_transforms(img_size, augment=False))

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=0, pin_memory=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=0, pin_memory=True)

    return train_loader, val_loader
```

```python
# ------------------------- STEP 8: TRAIN & VALIDATE FUNCTIONS -------------------------
def train_one_epoch(model, loader, optimizer, criterion, device):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    for images, labels in loader:
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)
    return running_loss / len(loader), correct / total

def validate(model, loader, criterion, device):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            running_loss += loss.item()
            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)
    return running_loss / len(loader), correct / total
```

```python
def objective(trial):
    # -------------------------
    # Hyperparameters
    # -------------------------
    backbone = trial.suggest_categorical("backbone", ["resnet18", "resnet50", "efficientnet_b0"])
    batch_size = trial.suggest_categorical("batch_size", [8, 16, 32])
    lr = trial.suggest_float("lr", 1e-5, 1e-3, log=True)
    img_size = trial.suggest_categorical("img_size", [224, 256])
    num_layers = trial.suggest_int("num_layers", 1, 3)
    d_model = trial.suggest_categorical("d_model", [128, 256])
    nhead = trial.suggest_categorical("nhead", [4, 8])
    loss_name = trial.suggest_categorical("loss", ["cross_entropy", "focal"])
    num_epochs = 3

    fold_list = ["fold_0", "fold_1", "fold_2"]
    fold_acc = []

    for i in range(len(fold_list)):
        val_fold = [os.path.join(TRAIN_BASE, fold_list[i])]
        train_folds = [os.path.join(TRAIN_BASE, f) for j, f in enumerate(fold_list) if j != i]

        # -------------------------
        # Data loaders
        # -------------------------
        train_loader, val_loader = get_fold_loaders(train_folds, val_fold, img_size, batch_size)

        # -------------------------
        # Model
        # -------------------------
        model = CNNTransformerClassifier(
            backbone_name=backbone,
            num_classes=2,
            d_model=d_model,
            nhead=nhead,
            num_layers=num_layers,
            pretrained=True
        ).to(device)

        # -------------------------
        # Freeze backbone except last block
        # -------------------------
        for p in model.backbone.parameters():
            p.requires_grad = False

        if "efficientnet" in backbone:
            # EfficientNet features is nn.Sequential
            for p in list(model.backbone.children())[-1].parameters():
                p.requires_grad = True
        elif "resnet" in backbone:
            for p in model.backbone.layer4.parameters():
                p.requires_grad = True

        # -------------------------
        # Loss, Optimizer, Scheduler
        # -------------------------
        criterion = nn.CrossEntropyLoss() if loss_name == "cross_entropy" else FocalLoss(alpha=1, gamma=2)
        optimizer = optim.AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=lr, weight_decay=1e-4)
        scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode="min", factor=0.5, patience=2)

        # -------------------------
        # Training with pruning
        # -------------------------
        best_acc = 0
        for epoch in range(num_epochs):
            train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, criterion, device)
            val_loss, val_acc = validate(model, val_loader, criterion, device)
            scheduler.step(val_loss)

            # Update best accuracy
            if val_acc > best_acc:
                best_acc = val_acc

            # -------------------------
            # Print per epoch
            # -------------------------
            print(f"Fold {i+1}, Epoch {epoch+1}/{num_epochs} -> train_acc: {train_acc:.4f}, val_acc: {val_acc:.4f}")

            # -------------------------
            # Pruning
            # -------------------------
            trial.report(best_acc, step=epoch)
            if trial.should_prune():
                raise optuna.exceptions.TrialPruned()

        fold_acc.append(best_acc)

    # -------------------------
    # Average accuracy across folds
    # -------------------------
    return sum(fold_acc) / len(fold_acc)
```

```python
# ------------------------- STEP 11: RUN OPTUNA -------------------------
study = optuna.create_study(direction="maximize", pruner=optuna.pruners.MedianPruner(n_warmup_steps=1))
study.optimize(objective, n_trials=20)

print("Best trial parameters:")
print(study.best_trial.params)
```

```python
import torch
import torch.nn as nn
from torchvision import models

class CNNTransformerClassifier(nn.Module):
    def __init__(self, backbone_name="resnet18", num_classes=2,
                 d_model=256, nhead=8, num_layers=2, pretrained=True):
        super().__init__()

        # -------- BACKBONE --------
        if backbone_name == "resnet18":
            self.backbone = models.resnet18(pretrained=pretrained)
            feature_dim = self.backbone.fc.in_features
            self.backbone.fc = nn.Identity()

        elif backbone_name == "resnet50":
            self.backbone = models.resnet50(pretrained=pretrained)
            feature_dim = self.backbone.fc.in_features
            self.backbone.fc = nn.Identity()

        elif backbone_name == "efficientnet_b0":
            self.backbone = models.efficientnet_b0(pretrained=pretrained)
            feature_dim = self.backbone.classifier[1].in_features
            self.backbone.classifier = nn.Identity()

        else:
            raise ValueError("Unsupported backbone")

        # -------- TRANSFORMER --------
        self.feature_proj = nn.Linear(feature_dim, d_model)

        self.transformer = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(
                d_model=d_model, nhead=nhead, batch_first=True
            ),
            num_layers=num_layers
        )

        # -------- CLASSIFIER --------
        self.classifier = nn.Sequential(
            nn.Linear(d_model, 256),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        x = self.backbone(x)                 # [B, C]
        x = self.feature_proj(x)             # [B, d_model]
        x = x.unsqueeze(1)                   # [B, 1, d_model]
        x = self.transformer(x)              # [B, 1, d_model]
        x = x.squeeze(1)                     # [B, d_model]
        return self.classifier(x)
```

```python
import torch.nn.functional as F

def ensemble_predict(models, images):
    """
    models: list of trained models
    images: tensor [B, C, H, W]
    """
    probs = []

    with torch.no_grad():
        for model in models:
            outputs = model(images)
            probs.append(F.softmax(outputs, dim=1))

    avg_probs = torch.mean(torch.stack(probs), dim=0)
    preds = torch.argmax(avg_probs, dim=1)

    return preds, avg_probs
```
